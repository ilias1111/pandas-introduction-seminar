# ***Questions***

**Q1:** Who is the youngest and the oldest athlete ever?

```
df.sort_values(by=['Age']).head(1)
```

```
age = df['Age'].min()
```

```
df.query("Age == @age")
```

**Q2:** Who is the youngest and the oldest athlete ever that has a gold metal?

```
df.query('Medal=="Gold"',inplace=False).sort_values(by=['Age'])[:1]
```

```
df['Age'][df['Medal'] == 'Gold'].min()
```

**Q3:** Who is the youngest and the oldest athlete ever that has a medal?

```
%%timeit
df.query('Medal=="Gold" or Medal=="Silver" or Medal=="Bronze"',inplace=False).sort_values(by=['Age']).iloc[[0]]
```

```
%%timeit
df[~df['Medal'].isna()].sort_values(by='Age', ascending=False).head(1)
```

**Q4:** Display all the judo athletes that have a gold medal.

```
df.loc[(df.Sport=="Judo") & (df.Medal=='Gold')]
```

```
df.query('(Sport=="Judo") and (Medal=="Gold")')
```

**Q5:** Display all the athletes that compete with the GRE in 2004

```
df.loc[(df.Year==2004)&(df.NOC=="GRE")]
```

**Q6:** Display all the gold medals GRE has ever won

```
df.loc[(df.NOC=="GRE") & (~df['Medal'].isna())]
```

**Q7:** Display all the athletes named "Alexandros"

```
df.set_index('Name').filter(like='Alexandros',axis=0)
```

**Q8:** What is the male/female ratio?

```
df.drop_duplicates(subset="ID")['Sex'].value_counts()
```

```
df['Sex'].value_counts()
```

```
df.query("Sex=='M'").shape[0]/df.query("Sex=='F'").shape[0]
```

**Q9:** Display the top 10 athletes in terms of medals

```
df.groupby(by=["ID","Name"])["Medal"].count().reset_index().sort_values(by='Medal', ascending=False).head(10)
```

**Q10:** Display the top 10 NOC in terms of gold medals

```
df[df.Medal=='Gold'].groupby(by=['NOC'])['Medal'].count().reset_index().sort_values(by='Medal', ascending=False).head(10)
```

**Q11:** Display the youngest and the oldest competitor in each sport

```
df.groupby('Sport').agg(
    Count=pd.NamedAgg(column='Year', aggfunc='count'),
    min_Age=pd.NamedAgg(column='Age', aggfunc=min),
    max_Age=pd.NamedAgg(column='Age', aggfunc=max)
)
```

```
df.groupby(['Sport']).Age.agg([min,max])
```
